```python
import pandas as pd
import os
import numpy as np
!pip install openpyxl

# Define file path
excel_file = os.path.join(r'ImportFile.xlsx')

# Import data
GDP_dat = pd.read_excel(excel_file, sheet_name='GDP')
Emp_dat = pd.read_excel(excel_file, sheet_name='Emp')
Cons_dat = pd.read_excel(excel_file, sheet_name='Cons')
IPExp_dat = pd.read_excel(excel_file, sheet_name='IPExp')

# Convert 'Date' columns to datetime
GDP_dat['Date'] = pd.to_datetime(GDP_dat['Date'])
Emp_dat['Date'] = pd.to_datetime(Emp_dat['Date'])
Cons_dat['Date'] = pd.to_datetime(Cons_dat['Date'])
IPExp_dat['Date'] = pd.to_datetime(IPExp_dat['Date'])

# Merge all datasets on 'Date'
data = GDP_dat[['Date', 'GDP']].merge(Emp_dat, on='Date', how='left')
data = data.merge(Cons_dat, on='Date', how='left')
data = data.merge(IPExp_dat, on='Date', how='left')

# Set 'Date' as index if desired
data.set_index('Date', inplace=True)

# Check data
print(data.head())
```

```python
data = data[~data.index.duplicated(keep='first')]
data.head()
```

```python
from statsmodels.tsa.stattools import adfuller

def test_stationarity(series, signif=0.05):
    result = adfuller(series, autolag='AIC')
    print(f"ADF Statistic: {result[0]}")
    print(f"p-value: {result[1]}")
    if result[1] < signif:
        print("Series is stationary")
    else:
        print("Series is not stationary")

for col in data.columns:
    print(f"Testing stationarity for {col}")
    test_stationarity(data[col])
```

```python
# First differencing of non-stationary series
data_diff = data.copy()  # Create a copy to store the differenced data
non_stationary_columns = ['GDP', 'U', 'Claims', 'Sentiment', 'Exp']  # List non-stationary columns based on ADF test results

for col in non_stationary_columns:
    data_diff[col] = data[col].diff().dropna()  # Apply differencing

# Drop the first row after differencing (it will be NaN)
data_diff = data_diff.dropna()

# Check stationarity again
for col in data_diff.columns:
    print(f"Re-testing stationarity for {col}")
    test_stationarity(data_diff[col])
```

```python
import matplotlib.pyplot as plt
from statsmodels.tsa.api import VAR

# Fit VAR Model on the differenced data
model_diff = VAR(data_diff)
results_diff = model_diff.fit(maxlags=2, ic='aic')  # Using AIC to select the best lag length
print(results_diff.summary())

# Perform Impulse Response Analysis with enhanced global settings
plt.figure(figsize=(14, 10))  # Set a larger figure size globally for all subplots
irf_diff = results_diff.irf(10)  # Analyzing 10 periods
irf_diff.plot()
plt.suptitle('Impulse Response Functions', fontsize=16, fontweight='bold')  # Set a title with a larger font
plt.show()

# Forecast Error Variance Decomposition (FEVD) with enhanced settings
plt.figure(figsize=(14, 10))
fevd_diff = results_diff.fevd(10)  # Up to 10 periods ahead
fevd_diff.plot()
plt.suptitle('Forecast Error Variance Decomposition (FEVD)', fontsize=16, fontweight='bold')
plt.show()
#need to set axis lables to south...

```

```python
irf = results_diff.irf(10)
fig, axes = plt.subplots(nrows=1, ncols=7, figsize=(20, 5), constrained_layout=True)
endog_names = results_diff.model.endog_names

for i in range(7):
    responses = irf.orth_irfs[:, i, 0]
    axes[i].plot(responses, label=f'Response to {endog_names[i]}')
    axes[i].axhline(0, color='black', linestyle='--', linewidth=0.5)
    axes[i].set_title(f'Response to {endog_names[i]}', fontsize=10)
    axes[i].set_xlabel('Periods', fontsize=8)
    axes[i].set_ylabel('Response', fontsize=8)
    axes[i].legend(fontsize=8)

plt.suptitle('GDP Responses in IRFs', fontsize=16, fontweight='bold')
plt.show()
```

```python
fevd = results_diff.fevd(10)
endog_names = results_diff.model.endog_names  # Get the names of the variables

# Set up the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Number of periods to display
periods = fevd.decomp.shape[0]

# Base of the bar (starts with zero)
base = None

# Colors for clarity
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink']

# Iterate through each variable and stack
for i in range(len(endog_names)):
    # Values to plot
    values = fevd.decomp[:, 0, i]
    ax.bar(range(periods), values, bottom=base, color=colors[i % len(colors)], edgecolor='white', label=f'Impact of {endog_names[i]}')
    # Update base for the next stack
    base = values if base is None else base + values

ax.set_title('Stacked FEVD for GDP')
ax.set_xlabel('Periods')
ax.set_ylabel('Fraction of Forecast Error Variance')
ax.legend(title="Variable Impact")
plt.xticks(range(periods), [str(x) for x in range(periods)])  # Ensure ticks correspond to periods
plt.show()
```

```python
gdp_results = results_diff.params['GDP'].to_frame('Coefficient')  # Extract coefficients
gdp_results['Std. Error'] = results_diff.stderr['GDP']  # Standard errors
gdp_results['t-stat'] = results_diff.tvalues['GDP']  # t-statistics
gdp_results['p-value'] = results_diff.pvalues['GDP']  # p-values

df_gdp = pd.DataFrame(gdp_results)

df_gdp.to_excel("GDP_VAR_Results.xlsx", sheet_name='GDP Analysis')

```

```python
residuals = results_diff.resid  
correlation_matrix = pd.DataFrame(residuals).corr()  

correlation_matrix.to_excel("Correlation_Matrix_of_Residuals.xlsx", sheet_name='Correlation Matrix')
```

```python
# Compute the FEVD for a specified number of periods
fevd = results_diff.fevd(10)

# Get the names of the variables from the model
endog_names = results_diff.model.endog_names

# Initialize a dictionary to hold the data
fevd_data = {}

# Extract FEVD values for each variable and store them in the dictionary
for i, name in enumerate(endog_names):
    fevd_data[f'Impact on GDP by {name}'] = fevd.decomp[:, 0, i]

# Convert the dictionary into a pandas DataFrame
fevd_df = pd.DataFrame(fevd_data, index=[f'Period {i+1}' for i in range(fevd.decomp.shape[0])])

fevd_df.to_excel("FEVD_Data.xlsx", sheet_name='FEVD Results')

```

```python
import scipy.stats as stats

# Extracting the parameters (coefficients), t-values, and we calculate p-values
coefs = results_diff.params
tvalues = results_diff.tvalues
pvalues = pd.DataFrame(2 * (1 - stats.norm.cdf(np.abs(tvalues))), columns=coefs.columns, index=coefs.index)  # Calculate p-values

# Creating a DataFrame to display results
results_df = pd.DataFrame({
    'Coefficient': coefs.stack(),
    'T-value': tvalues.stack(),
    'P-value': pvalues.stack()
})

# Function to apply conditional formatting for significance in DataFrame
def highlight_significant(val):
    color = 'red' if val < 0.05 else 'black'
    return f'color: {color}'

# Applying styling to DataFrame
styled_results = results_df.style.applymap(highlight_significant, subset=['P-value'])
styled_results
```
